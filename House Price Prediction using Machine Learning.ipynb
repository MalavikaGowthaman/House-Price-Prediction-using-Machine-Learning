```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
# import data
data = pd.read_csv('Housing.csv')
data
```

```python
# to get info
data.info()
```

```python
# drop the unwanted col
data = data.drop('stories',axis = 1)
data
```

```python
drop_col = ['prefarea','guestroom']
data = data.drop(drop_col,axis =1)
```

```python
data
```

```python
data.isna().sum()
```

```python
data= pd.get_dummies(data,drop_first = True)
data =data.astype(int)
data
```

```python
data.rename(columns={'furnishingstatus_semi-furnished':'furnishingstatus_semi_furnished'}, inplace =True)
data
```

```python
# to split input and output
independent = data[['area', 'bedrooms', 'bathrooms', 'parking', 'mainroad_yes',
       'basement_yes', 'hotwaterheating_yes', 'airconditioning_yes',
       'furnishingstatus_semi_furnished', 'furnishingstatus_unfurnished']]
dependent = data[['price']]
```

```python
# split data into training and  testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
x_train,x_test,y_train,y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor_f = LinearRegression()
regressor_f.fit(x_train,y_train)
```

```python
weight = regressor_f.coef_
bias = regressor_f.intercept_
print("The weight :",weight, "and the bias: ",bias)
```

```python
test_pred = regressor_f.predict(x_test)
test_pred
```

```python
from sklearn.metrics import r2_score
r2_score_value = r2_score(y_test,test_pred)
r2_score_value
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel ='linear')
regressor.fit(x_train,y_train)
```

```python
test_pred_svm = regressor.predict(x_test)
test_pred_svm
```

```python
from sklearn.metrics import r2_score
r2_score_value_svm = r2_score(y_test,test_pred_svm)
r2_score_value_svm
```

```python

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state= 0)
regressor.fit(x_train,y_train)
```

```python
test_pred_dt = regressor.predict(x_test)
test_pred_dt
```

```python
from sklearn.metrics import r2_score
r2_score_dt = r2_score(y_test,test_pred_dt)
r2_score_dt
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x_train,y_train)

```

```python
test_pred_rf = regressor.predict(x_test)
test_pred_rf
```

```python
from sklearn.metrics import r2_score
r2_score_value_rf = r2_score(y_test,test_pred_rf)
r2_score_value_rf
```
